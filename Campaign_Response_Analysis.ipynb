```python
marketing_cost_df = pd.read_csv('/content/AuraModeMarketingCostROIDataset 1.csv')
marketing_cost_df.head()
```

```python
# Analisis Campaign Response berdasarkan kedua dataset

# 1. Menghitung total biaya pemasaran, pendapatan, dan ROI (Revenue of Invesment) Rata-rata
total_marketing_cost = marketing_cost_df["Total_Marketing_Cost"].sum()
total_revenue = marketing_cost_df["Revenue_Generated"].sum()
average_roi = marketing_cost_df["ROI"].mean()

# 2. Menghitung Rata-rata CPA (Cost Per Acquisition)
average_cpa = marketing_cost_df["CPA"].mean()

# 3. Rata-rata Konversi pada Funnel Pembelian
campaign_df["Click_to_Cart_Conversion"] = (campaign_df["Add_to_Cart"] / campaign_df["Ad_Click"]).replace([float('inf'), -float('inf')], 0) * 100
campaign_df["Cart_to_Checkout_Conversion"] = (campaign_df["Checkout_Started"] / campaign_df["Add_to_Cart"]).replace([float('inf'), -float('inf')], 0) * 100
campaign_df["Checkout_to_Purchase_Conversion"] = (campaign_df["Purchase_Completed"] / campaign_df["Checkout_Started"]).replace([float('inf'), -float('inf')], 0) * 100

average_conversion_rates = {
    "Click to Cart Conversion (%)" : campaign_df["Click_to_Cart_Conversion"].mean(),
    "Cart to Checkout Conversion (%)" : campaign_df["Cart_to_Checkout_Conversion"].mean(),
    "Checkout to Purchase Conversion (%)" : campaign_df["Checkout_to_Purchase_Conversion"].mean()

}


# 4. Rata - rata bounce rate & Engagement rate
average_bounce_rate = campaign_df["Bounce_Rate"].mean()
average_engagement_rate = campaign_df["Engagement_Rate"].mean()

# 5. Performa berdasarkan platform
platform_performance = campaign_df.groupby("Platform").agg(
    Clicks = ("Ad_Click", "sum"),
    Add_to_Cart = ("Add_to_Cart", "sum"),
    Checkout_Started = ("Checkout_Started", "sum"),
    Purchase_Completed = ("Purchase_Completed", "sum"),
    CTR = ("CTR", "mean"),
    Bounce_Rate = ("Bounce_Rate", "mean"),
    Engagement_Rate = ("Engagement_Rate", "mean")
).reset_index()

# 6. ROI berdasarkan Platform
roi_per_platform = marketing_cost_df.groupby("Platform").agg(
    Total_Spend = ("Total_Marketing_Cost", "sum"),
    Total_Revenue = ("Revenue_Generated", "sum"),
    Average_ROI = ("ROI", "mean"),
    Average_CPA = ("CPA", "mean")
).reset_index()

# Menampilkan hasil analisis dalam tabel interaktif
# Mengganti tools.display_dataframe_to_user dengan display
print("Campaign Performance by Platform:")
display(platform_performance)
print("\nROI Analysis by Platform")
display(roi_per_platform)

total_marketing_cost, total_revenue, average_roi, average_cpa, average_conversion_rates, average_bounce_rate, average_engagement_rate
```

**2. Seberapa efektif influencer marketing dalam meningkatkan penjualan?**

```python
# menghitung total pembelian berdasarkan tipe influencer
influencer_sales = campaign_df.groupby("Influencer_Type")["Purchase_Completed"].sum()

# menghitung total jumlah pelanggan per kategori influencer
influencer_total_users = campaign_df.groupby("Influencer_Type")["Customer_ID"].count()

# menghitung purchase rate (persentase pelanggan yang melakukan pembelian)
influencer_purchase_rate = (influencer_sales / influencer_total_users) * 100

# menghitung rata-rata engagement rate per kategori influencer
influencer_engagement_rate = campaign_df.groupby("Influencer_Type")["Engagement_Rate"].mean()

# menggabungkan hasil ke dalam satu dataframe
influencer_effectiveness_df = pd.DataFrame({
    "Total_Purchase_Completed" : influencer_sales,
    "Total_Users" : influencer_total_users,
    "Purchase_Rate (%)" : influencer_purchase_rate,
    "AVG_Engagement_Rate" : influencer_engagement_rate
})

# menampilkan hasil kepada pengguna
display(influencer_effectiveness_df)
```

**1. Nano Influencers Lebih Efektif dalam Meningkatkan Konversi**

* Purchase Rate tertinggi berasal dari Nano Influencers (32.52%), lebih tinggi dibandingkan Macro (31.06%) dan Micro (30.62%).
* Hal ini menunjukkan bahwa pelanggan lebih percaya pada influencer dengan komunitas kecil tetapi lebih terhubung secara personal.

**2. Tidak Menggunakan Influencer Masih Cukup Efektif**

* Kategori "None" (tanpa influencer) memiliki purchase rate 28.84%, yang masih cukup tinggi dibandingkan dengan penggunaan influencer.
* Ini menunjukkan bahwa pelanggan tetap membeli meskipun tidak dipengaruhi oleh influencer.

**3. Engagement Rate Tidak Selalu Berarti Konversi Tinggi**
* Macro Influencers memiliki engagement rate tertinggi (3.41%), tetapi purchase rate mereka tidak tertinggi.
* Artinya, banyak orang berinteraksi dengan postingan mereka, tetapi tidak selalu melakukan pembelian.

**4. Efisiensi Biaya Influencer Perlu Dievaluasi**
* Influencer Marketing mungkin tidak selalu efisien dalam hal biaya, terutama jika CPA (Cost per Acquisition) dari influencer terlalu tinggi dibandingkan metode pemasaran lain.
* Jika biaya per akuisisi pelanggan dari influencer lebih mahal daripada iklan berbayar (Facebook Ads, Instagram Ads), maka sebaiknya anggaran influencer dikurangi.

**3. Apakah ada kendala dalam website dan proses checkout?**

```python
# analisis kendala dalam website dan proses checkout

# menghitung bounce rate rata-rata
avg_bounce_rate = campaign_df["Bounce_Rate"].mean()

# menghitung rata-rata waktu checkout dalam detik
avg_checkout_time = campaign_df["Checkout_Time_Sec"].mean()

# menghitung jumlah pengguna yang memulai checkout tetapi tidak menyelesaikan pembelian
checkout_started = campaign_df["Checkout_Started"].sum()
purchase_completed = campaign_df["Purchase_Completed"].sum()
dropoff_rate_checkout = ((checkout_started - purchase_completed) / checkout_started) * 100

# membuat dataframe hasil analisis
checkout_analysis_df = pd.DataFrame({
    "Metric" : ["Average Bounce Rate (%)", "Average Checkout Time (Second)", "Drop-off Rate During Checkout (%)"],
    "Value" : [avg_bounce_rate, avg_checkout_time, dropoff_rate_checkout]
})

# menampilkan hasil kepada pengguna
display(checkout_analysis_df)
```

**1. Bounce Rate Tinggi (65.12%)**

Ini menunjukkan bahwa banyak pengunjung meninggalkan website tanpa melakukan interaksi lebih lanjut.

Kemungkinan penyebab:
* UI/UX website kurang menarik atau sulit dinavigasi.
* Halaman produk atau deskripsi tidak cukup meyakinkan pelanggan.
* Kecepatan loading website mungkin lambat.

**2. Waktu Checkout Relatif Lama (167.52 detik / ±2.8 menit)**

Ini menunjukkan bahwa pelanggan menghabiskan waktu yang cukup lama di halaman checkout.

Kemungkinan penyebab:
* Proses checkout terlalu rumit atau membutuhkan banyak langkah.
* Terlalu banyak informasi yang harus diisi oleh pelanggan.
* Kurangnya metode pembayaran yang sesuai dengan preferensi pelanggan.

**3. Tingkat Drop-off Saat Checkout 22.05%**

Ini berarti 1 dari 5 pelanggan yang memulai checkout tidak menyelesaikannya.

Kemungkinan penyebab:
* Biaya tambahan (ongkir, pajak) muncul di akhir proses dan membuat pelanggan membatalkan pembelian.
* Tidak ada insentif (seperti diskon atau cashback) yang mendorong pelanggan untuk menyelesaikan pembelian.
* Kurangnya kepercayaan pelanggan terhadap keamanan pembayaran.

**4. Bagaimana pengaruh strategi harga dan promo terhadap pembelian?**

```python
# Analisis Pengaruh Strategi Harga dan Promo terhadap Pembelian

# Menghitung rata-rata jumlah pembelian untuk pelanggan yang mendapatkan diskon vs yang tidak
discount_analysis = campaign_df.groupby("Discount_Applied")["Purchase_Completed"].agg(["count", "sum"])
discount_analysis["Purchase_Rate (%)"] = (discount_analysis["sum"] / discount_analysis["count"]) * 100

# Menghitung rata-rata nilai pembelian berdasarkan penerapan diskon
avg_purchase_amount = campaign_df.groupby("Discount_Applied")["Purchase_Amount"].mean()

# Menggabungkan hasil analisis
discount_analysis["Avg_Purchase_Amount"] = avg_purchase_amount

# Menampilkan hasil kepada pengguna
display(discount_analysis)
```

**1. Pelanggan Tanpa Diskon Memiliki Purchase Rate Lebih Tinggi (31.96%)**

* Pelanggan yang tidak menerima diskon memiliki purchase rate lebih tinggi (31.96%) dibandingkan pelanggan yang mendapatkan diskon (28.19%).
* Ini bisa menunjukkan bahwa diskon tidak selalu menjadi faktor utama dalam keputusan pembelian.

Kemungkinan penyebab:
* Produk yang ditawarkan memang sudah memiliki daya tarik kuat tanpa perlu diskon besar.
* Diskon yang diberikan mungkin tidak cukup menarik atau tidak dipromosikan dengan baik.

**2. Pelanggan yang Mendapat Diskon Membeli dengan Nilai Lebih Tinggi (Rp139.5K vs Rp111.3K)**

* Rata-rata nilai pembelian pelanggan yang mendapat diskon lebih tinggi (Rp139.5K) dibandingkan pelanggan tanpa diskon (Rp111.3K).
* Ini menunjukkan bahwa diskon bisa mendorong pelanggan untuk membeli dalam jumlah lebih besar atau produk dengan harga lebih tinggi.
